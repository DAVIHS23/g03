## df_climber

```python
df_climber = pd.read_csv('datasource/climber_df.csv')
df_climber.head(5)
```

```python
def calculate_bmi(row):
    weight_kg = row['weight']
    height_m = row['height']
    return weight_kg / (height_m/100 * height_m/100)

df_climber['bmi'] = df_climber.apply(calculate_bmi, axis=1).round(2)
```

```python
df_climber.dtypes
```

```python
df_climber.describe()
```

Aus dieser Zusammenfassung sehe ich, dass es bei folgenden Features ein Fehler besteht:
- year_first = hier gibt es falsche Werte (z.Bsp. 0). Diese werden entfernt.

Zuerst geben wir diese Datensätze aus:

```python
df_climber = df_climber.sort_values(by='year_first', ascending=True)
df_climber.head(5)
```

Es werden somit 3 Zeilen aus dem Dataset gelöscht.

```python
df_climber_cleaned = df_climber[df_climber['year_first'] >= 1991]
```

```python
df_climber_cleaned.head(5)
```

```python
print(df_climber_cleaned[df_climber_cleaned['year_first'] > df_climber_cleaned['year_last']])
```

Es sind somit 247 Zeilen, in welchen das Year-First grösser ist als das Year-Last. Diese werden entfernt.

```python
df_climber_cleaned = df_climber_cleaned[df_climber_cleaned['year_first'] <= df_climber_cleaned['year_last']]
df_climber_cleaned.describe()
```

```python
null_count = df_climber_cleaned.isna().sum()
print(null_count)
```

```python
df_climber_cleaned.head(5)
```

```python
columns_to_round = ['grades_first','grades_last','grades_max','grades_mean']
df_climber_cleaned[columns_to_round] = np.ceil(df_climber_cleaned[columns_to_round]).astype(int)
df_climber_cleaned_fra = df_climber_cleaned.merge(df_grades[['grade_id', 'grade_fra']], left_on='grades_mean', right_on='grade_id')
df_climber_cleaned_fra = df_climber_cleaned_fra.drop(columns='grade_id')
df_climber_cleaned_fra.head()
```

## df_routes

```python
df_routes = pd.read_csv('datasource/routes_rated.csv')
df_routes.head()
```

```python
df_routes.describe()
```

```python
null_count = df_routes.isna().sum()
print(null_count)
```

Folgende Anpassungen sollen im Dataset vorgenommen werden:
- country: Das Kürzel für das Land soll mit Grossbuchstaben enthalten sein, da die ISO-Tabelle ebenfalls Grossbuchstaben hat.
- grade_mean: Der Schwierigkeitsgrad der Routen soll abgerundet werden und anhand dieses wertes mit den Schwierigkeitsgrad gemäss Französischer Skala (grade_fra) ergänzt werden.

```python
df_routes['country'] = df_routes['country'].str.upper()
df_routes['grade_mean'] = np.ceil(df_routes['grade_mean']).astype(int)
df_routes_grade_fra = df_routes.merge(df_grades[['grade_id', 'grade_fra']], left_on='grade_mean', right_on='grade_id')
df_routes_grade_fra = df_routes_grade_fra.drop(columns='grade_id')
df_routes_grade_fra.head()
```

```python
# Show all countries, which ar in the routes-Dataset
df_routes_cities = pd.DataFrame({'countries': df_routes_grade_fra['country'].unique()})
df_routes_cities.head(5)
```

```python
routesCountries_not_in_isoCountries = df_routes_cities[~df_routes_cities['countries'].isin(df_isoCountry['alpha-3'])]
print(routesCountries_not_in_isoCountries)
```

In der Routen-Tabelle wird ROM geführt, aber diese Abkürzung gibt es gemäss ISO nicht. Wie heissen die Klettergebiete, welche mit ROM als Land angegeben werden?

```python
df_routes_grade_fra[df_routes_grade_fra['country']=='ROM'].head()
```

Gemäss Google ist Ampoita in Romania. Rumänien hat gemäss ISO jedoch den Code ROU. Somit wird das Land von ROM in ROU geändert.

```python
df_routes_grade_fra['country'] = df_routes_grade_fra['country'].replace('ROM','ROU')
df_routes_grade_fra[df_routes_grade_fra['country']=='ROU'].head()
```

```python
df_routes_fra_merged = df_routes_grade_fra.merge(df_isoCountry, left_on='country', right_on='alpha-3', how='left')
df_routes_fra_merged = df_routes_fra_merged.drop(columns='alpha-3')
df_routes_fra_merged.head()
```

```python
df_routes_fra_merged['grade_mean'] = np.ceil(df_routes_fra_merged['grade_mean']).astype(int)
df_routes_fra_merged.head()
```
